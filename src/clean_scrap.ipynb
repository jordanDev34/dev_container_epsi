I. Import & config

```python
import pandas as pd
```

II. Chargement fichier CSV

```python
df = pd.read_csv('numbeo_scraping.csv')
print(df.head())
```

III. Initialisation de la colonne Category'

```python
df['Category'] = None
```

IV. Définition des catégories via les lignes

```python
categories = {
    "Restaurants": (1, 9),  # Plage des lignes pour Restaurants
    "Markets": (10, 28),    # Plage des lignes pour Markets
    "Transportation": (30, 37),
    "Utilities (Monthly)": (39, 41),
    "Sports And Leisure": (43, 45),
    "Childcare": (47, 48),
    "Clothing And Shoes": (50, 53),
    "Rent Per Month": (55, 58),
    "Buy Apartment Price": (59, 61),
    "Salaries And Financing": (63, 65),
}


# Affecter les catégories en fonction des plages
for category, (start, end) in categories.items():
    df.loc[start:end, 'Category'] = category
```

V. Nettoyage des données (Edit ou sans données)

```python
df = df[df['0'] != 'Edit']  # Suppression des lignes avec 'Edit'
df = df.dropna(subset=['0', '1', '2'], how='all')  # Supprimer les lignes vides


# Colonnes renommées pour plus de clarté
df.columns = ['Description', 'Price', 'Range', 'Category']
```

VI. Exporter le fichier nettoyé

```python
# Exporter le fichier nettoyé
df.to_csv('numbeo_cleaned.csv', index=False, encoding='utf-8')
print("Les données nettoyées ont été exportées dans 'numbeo_cleaned.csv'.")
```
