# Cleaning cost of Living Date

I. Import & config

```python

import pandas as pd
```

II. Chargement du fichier CSV

```python

# Chargement du fichier CSV contenant les données brutes
df = pd.read_csv('scraped_cost_of_living_data.csv')
df.head()
```

III. Nettoyage des données

```python
# 1. Création d'une copie 
df_cleaned = df.copy()
```

```python
# 2. Suppression des symboles monétaires (par exemple €, $, £) dans la colonne 'Price'
df_cleaned.loc[:, 'Price'] = df_cleaned['Price'].replace({'€': '', '$': '', '£': ''}, regex=True)
```

```python
# 3. Conversion de la colonne 'Price' en numérique (float), en remplaçant les valeurs non convertibles par NaN => erreur ne fonctionne pas et supprime toutes les données
# df_cleaned['Price'] = pd.to_numeric(df_cleaned['Price'], errors='coerce')
```

```python
# 4. Suppression des lignes contenant des valeurs manquantes dans la colonne 'Price'
df_cleaned = df_cleaned.dropna(subset=['Price'])
```

```python
# 5. Affichage des 50 premières lignes des données nettoyées
df_cleaned.head(50)
```

IV. Sauvegarde des données nettoyées

```python

# Sauvegarde des données nettoyées dans un nouveau fichier CSV
df_cleaned.to_csv('cleaned_cost_of_living_data.csv', index=False, encoding='utf-8')

print("Les données nettoyées ont été enregistrées dans 'cleaned_cost_of_living_data.csv'.")
```
