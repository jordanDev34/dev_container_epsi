I. Import & config

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

II. Chargement fichier CSV

```python
df_urls = pd.read_csv('numbeo_country_urls.csv')
print(df_urls.head())
```

```python
# J'intialise une liste vide pour stocker les données
data = []
```

III. Je boucle sur chaque URL et je scrape les données de coût de la vie pour chaque pays


```python
for country_url in df_urls['Country_URL']:
    full_url = f"https://www.numbeo.com/cost-of-living/{country_url}"
    print(f"Scraping data for {full_url}")
    
    # J'envoie la requête à la page du pays
    response = requests.get(full_url)
    
    # Je vérifie si la requête a réussi
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # J'extrait toutes les lignes de la page
        rows = soup.find_all('tr')
        
        # Je boucle sur toutes les lignes et je collecte les données
        for row in rows:
            columns = row.find_all('td')
            if len(columns) > 1:
                item = columns[0].get_text(strip=True)
                cost = columns[1].get_text(strip=True)
                range_ = columns[2].get_text(strip=True) if len(columns) > 2 else ''
                data.append([country_url, item, cost, range_])
    else:
        print(f"Failed to retrieve data for {full_url}")
```

```python
# Je crée un DataFrame pour les données extraites
df_data = pd.DataFrame(data, columns=["Country_url", "Item", "Price", "Range"])
```

IV. Extraction des données en fichier CSV

```python
# Je sauvegarde les données dans un fichier CSV nommé 'scraped_cost_of_living_data.csv'
df_data.to_csv('scraped_cost_of_living_data.csv', index=False, encoding='utf-8')
print("Data has been saved to 'scraped_cost_of_living_data.csv'.")
```
