I. Import & config

```python
import pandas as pd
```

II. Chargement des données pour vérification

```python
df = pd.read_csv('numbeo_scraping.csv')
print(df.head())
```

III. Nettoyage des données

```python
df.dropna(inplace=True)
```

```python
df
```

IV. Renommer mes colonnes

```python
df.columns = ['Catégorie', 'Prix', 'Gamme']
```

V. Extraction des catégories principales

```python
def categorize(row):
    if 'Restaurant' in row['Catégorie']:
        return 'Restaurants'
    elif 'Marchés' in row['Catégorie'] or 'Lait' in row['Catégorie']:
        return 'Marchés'
    elif 'Transport' in row['Catégorie']:
        return 'Transport'
    else:
        return 'Autres'
```

VI. Ajout d'une colonne

```python
df['Type'] = df.apply(categorize, axis=1)
```

VII. Exportation des données nettoyées

```python
df.to_csv('numbeo_cleaned.csv', index=False, encoding='utf-8')
print("Les données nettoyées ont été exportées dans le fichier 'numbeo_cleaned.csv'.")
```
